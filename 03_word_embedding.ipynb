2. Word Embedding model을 이용하여 벡터화한다.
3. 입력된 문자열의 긍/부정을 판단한다. (유사도 활용)

### 데이터 로드

```python
# 01_preprocessing.pkl로 저장해둔 전처리 결과물 로드
import pickle

with open('./data/01_preprocessing.pkl', 'rb') as f:
    texts_gemini = pickle.load(f)

print(texts_gemini)
```

### FastText Model로 벡터화

```python
# FastText model 생성
from gensim.models import FastText

ft_model = FastText(
    sentences=texts_gemini,
    vector_size=100,
    window=5,
    min_count=5,
    sg=0
)
```

```python
# vocab 사이즈
print(ft_model.wv.vectors.shape)
```

### 입력된 문자열의 긍/부정 판단

```python
# 전처리 결과물의 문장별 단어들을 str 형태로 묶기
texts_gemini_str = []
for sentence in texts_gemini:
    texts_gemini_str.append(' '.join(word for word in sentence))

print(texts_gemini_str)
```

```python
# 긍/부정을 판단할 기준 문자열 생성
pos_sent = '가능 간단 감명 강력 강렬 강점 능력 도움 매끄러운 빠르게 빠른 성과 성능 유능 인기 정확 지능 지식 통찰 프리미엄 해결 효율 희망'
neg_sent = '경고 경쟁 공격 보복 부족 비판 실망 조롱'
```

```python
# 긍정 기준 문자열 23개, 부정 기준 문자열 8개로 유사도 계산
pos_count, neg_count = 0, 0
pos_similarity = []
neg_similarity = []

for str in texts_gemini_str:
    pos_similarity.append(ft_model.wv.similarity(str, pos_sent))    # 전처리 데이터 내 문장과 긍정 문자열과의 유사도 계산
    neg_similarity.append(ft_model.wv.similarity(str, neg_sent))    # 전처리 데이터 내 문장과 부정 문자열과의 유사도 계산

    if ft_model.wv.similarity(str, pos_sent) > ft_model.wv.similarity(str, neg_sent):
        pos_count += 1      # 긍정 유사도가 더 높으면
    else:
        neg_count += 1      # 부정 유사도가 더 높으면

pos_similarity = sorted(pos_similarity, reverse=True)
neg_similarity = sorted(neg_similarity, reverse=True)

if pos_count > neg_count:
    print(f'총 {len(texts_gemini)}개의 문자열 중 과반수 이상인 {pos_count}개의 문자열이 "긍정" 성향이었습니다.')
    print('최고 similarity 상위 5개 :', pos_similarity[0], pos_similarity[1], pos_similarity[2], pos_similarity[3], pos_similarity[4])
else:
    print(f'총 {len(texts_gemini)}개의 문자열 중 과반수 이상인 {neg_count}개의 문자열이 "부정" 성향이었습니다.')
    print('최고 similarity 상위 5개 :', neg_similarity[0], neg_similarity[1], neg_similarity[2], neg_similarity[3], neg_similarity[4])
```

```python
# 긍/부정을 판단할 기준 문자열 재생성 - 긍정 문자열 개수 8개로 맞춰줌
pos_sent = '감명 강력 강렬 능력 유능 정확 지능 효율'
neg_sent = '경고 경쟁 공격 보복 부족 비판 실망 조롱'
```

```python
# 긍정 기준 문자열 8개, 부정 기준 문자열 8개로 유사도 계산
pos_count, neg_count = 0, 0
pos_similarity = []
neg_similarity = []

for str in texts_gemini_str:
    pos_similarity.append(ft_model.wv.similarity(str, pos_sent))
    neg_similarity.append(ft_model.wv.similarity(str, neg_sent))

    if ft_model.wv.similarity(str, pos_sent) > ft_model.wv.similarity(str, neg_sent):
        pos_count += 1
    else:
        neg_count += 1

pos_similarity = sorted(pos_similarity, reverse=True)
neg_similarity = sorted(neg_similarity, reverse=True)

if pos_count > neg_count:
    print(f'총 {len(texts_gemini)}개의 문자열 중 과반수 이상인 {pos_count}개의 문자열이 "긍정" 성향이었습니다.')
    print('최고 similarity 상위 5개 :', pos_similarity[0], pos_similarity[1], pos_similarity[2], pos_similarity[3], pos_similarity[4])
else:
    print(f'총 {len(texts_gemini)}개의 문자열 중 과반수 이상인 {neg_count}개의 문자열이 "부정" 성향이었습니다.')
    print('최고 similarity 상위 5개 :', neg_similarity[0], neg_similarity[1], neg_similarity[2], neg_similarity[3], neg_similarity[4])
```

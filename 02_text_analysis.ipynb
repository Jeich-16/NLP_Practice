# BoW 기반 텍스트 분석 연습

1. 적절한 데이터셋을 찾거나 생성하고, 적절한 전처리를 진행한다. [01_preprocessing.ipynb]
2. TF-IDF Vectorizer를 이용하여 벡터화한다.
3. Cosine Similarity를 계산하여 입력된 문자열의 긍/부정을 판단한다.

### 데이터 로드

```python
# 01_preprocessing.pkl로 저장해둔 전처리 결과물 로드
import pickle

with open('./data/01_preprocessing.pkl', 'rb') as f:
    texts_gemini_clean = pickle.load(f)

print(texts_gemini_clean)
```

```python
# 문장별 단어들을 str 형태로 묶기
texts_gemini = []
for sentence in texts_gemini_clean:
    texts_gemini.append(' '.join(word for word in sentence))

print(texts_gemini)
```

### TF-IDF Vectorizer로 벡터화

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
# tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, 2))
texts_gemini_vecs = tfidf_vectorizer.fit_transform(texts_gemini)

display(texts_gemini_vecs.toarray())
tfidf_vectorizer.get_feature_names_out()
```

### 입력된 문자열의 긍/부정 판단

```python
# 긍/부정을 판단할 기준 문자열 생성
pos_sent = '좋다 괜찮다 기쁘다 만족 추천 행복'
neg_sent = '나쁘다 별로다 슬프다 불만족 비추천 불행'

# 전처리한 데이터에 긍/부정 문자열 합친 후 벡터화 재진행
texts_all = texts_gemini + [pos_sent] + [neg_sent]

tfidf_vectorizer = TfidfVectorizer()
# tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, 2))
texts_all_vecs = tfidf_vectorizer.fit_transform(texts_all)

# 벡터 분리 : 전처리 데이터 / 긍정 문자열 / 부정 문자열
texts_gemini_vecs = texts_all_vecs[:-2]
pos_sent_vec = texts_all_vecs[-2]
neg_sent_vec = texts_all_vecs[-1]
```

```python
# 긍정 <> 데이터 / 부정 <> 데이터 Cosine Similarity 계산
from sklearn.metrics.pairwise import cosine_similarity

pos_sim = cosine_similarity(pos_sent_vec, texts_gemini_vecs)
neg_sim = cosine_similarity(neg_sent_vec, texts_gemini_vecs)

print(pos_sim)
print(neg_sim)
# 전부 0으로 나옴. 빈도 기반인 TfidfVectorizer임을 고려할 때 긍정/부정 문자열의 단어들이 등장하지 않아서인듯.
```

### 긍/부정 기준 문자열 수정 후 재진행

```python
# 긍/부정을 판단할 기준 문자열 재생성
# - 해당 텍스트 데이터가 위키피디아 '제미나이' 검색 결과 페이지인 만큼, 주제에 어울리도록 재설정 필요
# - tfidf_vectorizer.get_feature_names_out()로 조회되는 단어들 중 기준 문자열에 넣을만한 단어들 선정

pos_sent = '가능 간단 감명 강력 강렬 강점 능력 도움 매끄러운 빠르게 빠른 성과 성능 유능 인기 정확 지능 지식 통찰 프리미엄 해결 효율 희망'
neg_sent = '경고 경쟁 공격 보복 부족 비판 실망 조롱'

# 전처리한 데이터에 긍/부정 문자열 합친 후 벡터화 재진행
texts_all = texts_gemini + [pos_sent] + [neg_sent]

tfidf_vectorizer = TfidfVectorizer()
texts_all_vecs = tfidf_vectorizer.fit_transform(texts_all)

# 벡터 분리 : 전처리 데이터 / 긍정 문자열 / 부정 문자열
texts_gemini_vecs = texts_all_vecs[:-2]
pos_sent_vec = texts_all_vecs[-2]
neg_sent_vec = texts_all_vecs[-1]
```

```python
# 긍정 <> 데이터 / 부정 <> 데이터 Cosine Similarity 계산
from sklearn.metrics.pairwise import cosine_similarity

pos_sim = cosine_similarity(pos_sent_vec, texts_gemini_vecs)
neg_sim = cosine_similarity(neg_sent_vec, texts_gemini_vecs)

print(pos_sim)
print(neg_sim)
```

### Cosine Similarity 계산 결과

```python
pos_sorted_idx = pos_sim.argsort()[:, ::-1]   # 내림차순 정렬
pos_sorted_idx = pos_sorted_idx.reshape(-1)

pos_sim = pos_sim.reshape(-1)
```

```python
# Cosine Similarity 분석 결과 - 긍정
print('텍스트 데이터 내 문자열 개수 :', len(texts_gemini), '개')
print('긍정 성향의 문자열 개수 :', sum(1 for i in pos_sim[pos_sorted_idx] if i!= 0), '개')
print('긍정 성향일 확률 :', round((sum(1 for i in pos_sim[pos_sorted_idx] if i!= 0)) / (len(texts_gemini)) * 100, 2), '%')
print('최고 Similarity 상위 5개 :', pos_sim[pos_sorted_idx][:5])
```

```python
neg_sorted_idx = neg_sim.argsort()[:, ::-1]   # 내림차순 정렬
neg_sorted_idx = neg_sorted_idx.reshape(-1)

neg_sim = neg_sim.reshape(-1)
```

```python
# Cosine Similarity 분석 결과 - 부정
print('텍스트 데이터 내 문자열 개수 :', len(texts_gemini), '개')
print('부정 성향의 문자열 개수 :', sum(1 for i in neg_sim[neg_sorted_idx] if i!= 0), '개')
print('부정 성향일 확률 :', round((sum(1 for i in neg_sim[neg_sorted_idx] if i!= 0)) / (len(texts_gemini)) * 100, 2), '%')
print('최고 Similarity 상위 5개 :', neg_sim[neg_sorted_idx][:5])
```

### 데이터 로드
https://ko.wikipedia.org/wiki/%EC%A0%9C%EB%AF%B8%EB%82%98%EC%9D%B4_(%EC%96%B8%EC%96%B4_%EB%AA%A8%EB%8D%B8)
- 위키피디아 '제미나이' 검색 결과

```python
# wikipedia_gemini.txt 파일 로드 및 출력
def load_wikipedia_gemini(filepath):
    with open(filepath, 'r', encoding='UTF-8') as f:
        return [line.strip() for line in f]                     # 텍스트 데이터를 한 줄 또는 단락별로 구분하여 가져옴
    
texts = load_wikipedia_gemini('./data/wikipedia_gemini.txt')
print('초기 요소 개수 :', len(texts))
texts
```

```python
# 불필요한 단락 간 공백 요소('') 삭제
texts_gemini = []
for text in texts:
    if text != '':
        texts_gemini.append(text)

print('공백 삭제 후 요소 개수 :', len(texts_gemini))
texts_gemini
```

### 문장, 단어 단위 토큰화

```python
# nltk를 활용한 문장 토큰화
import nltk
texts_gemini_sent_tokenize = []
count_sentences = 0

for text in texts_gemini:
    texts_gemini_sent_tokenize.append(nltk.sent_tokenize(text))

for sentence in texts_gemini_sent_tokenize:
    count_sentences += len(sentence)

print('sent_tokenize 적용 후 문장 개수 :', count_sentences)

texts_gemini_sent_tokenize
```

```python
# nltk를 활용한 단어 토큰화
texts_gemini_word_tokenize = []
count_words = 0

for text in texts_gemini:
    texts_gemini_word_tokenize.append(nltk.word_tokenize(text))

for word in texts_gemini_word_tokenize:
    count_words += len(word)

print('word_tokenize 적용 후 단어 개수 :', count_words)

texts_gemini_word_tokenize
```

```python
# kss를 활용한 문장 토큰화
import kss

count_sentences = 0

texts_gemini_split_sentences = kss.split_sentences(texts_gemini)

for sentence in texts_gemini_split_sentences:
    count_sentences += len(sentence)

print('split_sentences 적용 후 문장 개수 :', count_sentences)       # sent_tokenize 결과와 30개 차이 남... 각주 영향인 듯

texts_gemini_split_sentences
```

### 정제

```python
# 각주(ex: [1]) 삭제
import re
texts_gemini_remove_footnote = []

pattern = r'\[\d+\]'

for split in texts_gemini_split_sentences:
    for sentence in split:
        texts_gemini_remove_footnote.append(re.sub(pattern, '', sentence))

texts_gemini_remove_footnote
```

### 불용어 제거

```python
# 형태소 분석
from konlpy.tag import Okt

texts_gemini_morphs = []
okt = Okt()

for sentence in texts_gemini_remove_footnote:
    texts_gemini_morphs.append(okt.morphs(sentence))

texts_gemini_morphs
# 고유명사는 쪼개지면 안 되는데 쪼개짐...
```

```python
# !pip install customized_konlpy
# 형태소 목록에 고유명사 등록을 위해 'ckonlpy' 라이브러리 설치
```

```python
# 보호해야 할 고유명사 리스트를 만들어 등록
from ckonlpy.tag import Twitter

twt = Twitter()
nouns_customized = [
    '제미나이', '위키백과', '딥마인드', '멀티모달'
    ]                                           # 고유명사 리스트
for noun in nouns_customized:
    twt.add_dictionary(noun, 'Noun')            # 형태소 목록에 등록
```

```python
# 형태소 재분석
texts_gemini_morphs = []

for sentence in texts_gemini_remove_footnote:
    texts_gemini_morphs.append(twt.morphs(sentence))

texts_gemini_morphs
```

```python
# 불용어 제거
def load_ko_stopwords(filepath):
    with open(filepath, 'r', encoding='UTF-8') as f:
        return [line.strip() for line in f]    
ko_stopwords = load_ko_stopwords('./data/ko_stopwords.txt')

texts_gemini_clean = []
for sentence in texts_gemini_morphs:
    for token in sentence:
        if token not in ko_stopwords:
            texts_gemini_clean.append(token)

texts_gemini_clean
```

```python
# 특수문자 제거
texts_gemini_clean_ver2 = []
pattern = '[a-zA-Z가-힣0-9]+'

for text in texts_gemini_clean:
    if re.findall(pattern, text):
        texts_gemini_clean_ver2.append(re.findall(pattern, text))

texts_gemini_clean_ver2
```

```python
# 중첩 구조인 list를 단일 구조로 변경
texts_gemini_flatten = []

for word_list in texts_gemini_clean_ver2:
    for word in word_list:
        texts_gemini_flatten.append(word)

texts_gemini_flatten
```
